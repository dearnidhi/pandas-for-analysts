### How Pandas stores data internally
Pandas DataFrames are backed by **NumPy arrays** (or Arrow arrays in newer versions).  
Understanding this is critical for writing **fast code**.

```
DataFrame
  └── columns (Series)
        └── Block (numpy ndarray, contiguous in memory)
```
- Same-dtype columns share a single block → fast operations  
- Mixed dtypes = multiple blocks → slower  
- This is why `df.select_dtypes()` + keeping clean dtypes **matters**
